```python
import os
import config
import pandas as pd
from tqdm.auto import tqdm
import numpy as np
```

## Ellie's utterance

```python
total_dict = {}
len_total = 0
len_unique = 0

for f in tqdm(os.listdir(os.path.join(config.DATA_DIR, 'Transcription'))):
  transcription = pd.read_csv(os.path.join(config.DATA_DIR, 'Transcription', f))
  total_t = transcription.loc[transcription.speaker=='Ellie', 'value']
  for s in total_t:
    if s not in total_dict.keys():
      total_dict[s] = 0
    total_dict[s] += 1
```
